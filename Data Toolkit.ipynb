Q.21. Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing?
  - NumPy‚Äôs array slicing is generally much faster than slicing Python lists because of how they are implemented and stored in memory. Here are the key reasons:

üîπ 1. Contiguous Memory Layout
NumPy arrays store elements in a contiguous block of memory (like C arrays).

This allows slicing to create views on the original data without copying it.

Python lists are arrays of pointers to objects scattered in memory, so slicing involves copying references.

üîπ 2. View vs Copy
NumPy slicing returns a view of the original array (no new data is copied).

Python list slicing creates a new list, copying references to the objects.

This means NumPy slicing avoids the overhead of copying data.

üîπ 3. Homogeneous Data Types
NumPy arrays hold elements of the same data type, enabling efficient memory access and operations.

Python lists can hold heterogeneous objects, so operations are less optimized.

üîπ 4. Low-level Optimizations
NumPy is implemented in C and optimized for numerical computing.

Operations on NumPy arrays leverage vectorized operations and optimized loops.

Python lists use more generic and slower Python-level code.

```python
#Write a Pandas script to find the mean of a specific column in a DataFrame
import pandas as pd

data = {'A': [10, 20, 30, 40], 'B': [5, 15, 25, 35]}
df = pd.DataFrame(data)

mean_A = df['A'].mean()
print("Mean of column A:", mean_A)
```

```python
#How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(10, 4), columns=list('ABCD'))
corr = df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix Heatmap')
plt.show()
```
